```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, TensorDataset
import torch.functional as F
from torchvision import datasets
from torchsummary import summary
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = torch.tensor([
    [
        [
            [1, 2, 3, 4],
            [2, 3, 4, 5],
            [5, 6, 7, 8],
            [1, 3, 4, 5]
        ],
    ],
    [
        [
            [-1, 2, 3, -4],
            [2, -3, 4, 5],
            [-5, 6, -7, 8],
            [-1, -3, -4, -5]
        ]
    ]
],
    dtype=torch.float32)

X_train /= X_train.max()
y_train = torch.tensor([0, 1], dtype=torch.float32)
```

```python
model = nn.Sequential(
    nn.Conv2d(1, 1, 3),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Flatten(),
    nn.Linear(1, 1),
    nn.Sigmoid(),
)

criterion = nn.BCELoss()
opt = optim.Adam(model.parameters())
```

```python
trn_dataloader = DataLoader(TensorDataset(X_train, y_train), 2)
```

```python
epochs = 1000

for epoch in range(epochs):
    for x, y in trn_dataloader:
        _y = model(x)
        loss = criterion(_y, y.unsqueeze(1))

        opt.zero_grad()
        loss.backward()
        opt.step()

        print(loss)
```

```python
(conv_w, conv_b), (lin_w, lin_b) = [(layer.weight, layer.bias) for layer in model.children() if hasattr(layer, 'weight')]

in_img = X_train[1].squeeze(0)

output = torch.zeros([2, 2])
squeezed_conv_w = conv_w.squeeze(0).squeeze(0)

outRow = 0
outCol = 0

for i in range(output.shape[0]):
    for j in range(output.shape[1]):
        img_subset = in_img[i:i+3, j:j+3]
        output[i, j] = torch.sum(img_subset * squeezed_conv_w) + conv_b

output.clamp_min_(0)
output = output.max()

output = (output * lin_w + lin_b).squeeze(0)
torch.sigmoid(output)
```
